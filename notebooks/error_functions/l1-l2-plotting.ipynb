Trying to understand the L1 vs L2 difference:
    
    L1 norm for vector V = [x, y] is |x| + |y|
    L2 norm for vector V = [x, y] is x^2 + y^2

```python
import numpy as np

def L1(V):
    n = len(V)
    l1 = 0.0
    for i in xrange(0,n):
        l1 += abs(V[i])
    return l1

def L2(V):
    n = len(V)
    l1 = 0.0
    for i in xrange(0,n):
        l1 += V[i]**2
    return l1
```

```python
n = 100 
X = np.linspace(-2,2,n)
Y = np.linspace(-2,2,n)
L1_vals = np.zeros_like(X)
L2_vals = np.zeros_like(X)
for i in xrange(0, n):
    V = [X[i], Y[i]]
    L1_vals[i] = L1(V)
    L2_vals[i] = L2(V)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, L1_vals)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('L1')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X, Y, L2_vals)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('L2')
```
